```python
from astar_core import *
```

```python
def run_test(nome, grid, tunnels):
    print(f"\n===== {nome} =====")

    start, goal = find_positions(grid)

    for h in [h1, h2, h3]:
        result = a_star(grid, start, goal, h, tunnels)

        print(f"\nHeurística: {h.__name__}")

        if result:
            print(f"Custo: {result['cost']}")
            print(f"Tempo: {result['time']:.6f}s")
            print(f"Visitados: {result['visited']}")
            print(f"Fronteira máxima: {result['max_frontier']}")
            print(f"Caminho: {result['path']}")
        else:
            print("Sem solução")
```

```python
grid_easy1 = [
    ['P', '.', '.', 'D']
]
tunnels_easy1 = {}

run_test("Fácil 1 - Direto", grid_easy1, tunnels_easy1)
```

```python
grid_easy2 = [
    ['P', '.', '.'],
    ['.', '.', 'D']
]
tunnels_easy2 = {}

run_test("Fácil 2 - Aberto", grid_easy2, tunnels_easy2)
```

```python
grid_medium1 = [
    ['P', 'S', 'S', 'S', 'D'],
    ['.', '.', '.', '.', '.']
]
tunnels_medium1 = {}

run_test("Médio 1 - Armadilha", grid_medium1, tunnels_medium1)
```

```python
grid_medium2 = [
    ['P', '.', 'S', '.', '.'],
    ['#', '#', 'S', '#', '.'],
    ['.', '.', '.', '.', 'D']
]
tunnels_medium2 = {}

run_test("Médio 2 - Obstáculos", grid_medium2, tunnels_medium2)
```

```python
grid_hard1 = [
    ['P', '.', '#', 'T'],
    ['#', 'S', '#', '.'],
    ['T', '.', '.', 'D']
]

tunnels_hard1 = {
    (0,3): (2,0),
    (2,0): (0,3)
}

run_test("Difícil 1 - Túnel obrigatório", grid_hard1, tunnels_hard1)
```

```python
grid_hard2 = [
    ['P', '.', '#', 'T', '.'],
    ['#', 'S', '#', '.', '.'],
    ['T', '.', '#', 'S', 'D'],
    ['.', '.', '.', '#', '.']
]

tunnels_hard2 = {
    (0,3): (2,0),
    (2,0): (0,3)
}

run_test("Difícil 2", grid_hard2, tunnels_hard2)
```
